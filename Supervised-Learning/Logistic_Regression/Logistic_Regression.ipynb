```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import  make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x,y=make_classification(n_samples=1000,n_features=20,n_classes=2,random_state=42)
```

```python
pd.DataFrame(x).head()
```

```python
pd.DataFrame(y).head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
logistic=LogisticRegression()
logistic.fit(x_train,y_train)
```

```python
y_pred=logistic.predict(x_test)
print(y_pred)
```

```python
logistic.predict_log_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_pred)
p_score=precision_score(y_test,y_pred)
r_score=recall_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
print("Accuracy Score:",score)
print("Precision Score:",p_score)
print("Recall Score:",r_score)
print("Confusion Matrix:\n",cm)
print("Classification Report:\n",cr)
```

## Hypertuning and Cross Validation

```python
model=LogisticRegression(max_iter=5000)
penalty=['l1','l2','elasticnet']
c_values=[100,10,1,0.1,0.001]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
Grid=GridSearchCV(estimator=model,param_grid=params,cv=cv,n_jobs=-1)
```

```python
Grid
```

```python
import warnings
warnings.filterwarnings('ignore')
Grid.fit(x_train,y_train)
```

```python
Grid.best_params_
```

```python
y_pred_grid=Grid.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_pred_grid)
p_score=precision_score(y_test,y_pred_grid)
r_score=recall_score(y_test,y_pred_grid)
cm=confusion_matrix(y_test,y_pred_grid)
cr=classification_report(y_test,y_pred_grid)
print("Accuracy Score:",score)
print("Precision Score:",p_score)
print("Recall Score:",r_score)
print("Confusion Matrix:\n",cm)
print("Classification Report:\n",cr)
```
